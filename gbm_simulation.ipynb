```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def simulate_gbm(S0, mu, sigma, T, dt, N):
    """
    模拟几何布朗运动
    S0: 初始价格
    mu: 期望收益率
    sigma: 波动率
    T: 总时间
    dt: 时间步长
    N: 模拟路径数
    """
    # 计算步数
    steps = int(T / dt)
    # 初始化价格矩阵
    S = np.zeros((N, steps))
    S[:, 0] = S0  # 设置初始价格

    for i in range(1, steps):
        # 生成标准正态分布随机数
        Z = np.random.normal(0, 1, N)
        # 欧拉方法更新价格
        S[:, i] = S[:, i - 1] * np.exp((mu - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * Z)

    return S

```

```python
# 参数设置
S0 = 100  # 初始价格
mu = 0.1  # 期望收益率
sigma = 0.2  # 波动率
T = 1  # 总时间（年）
dt = 0.01  # 时间步长
N = 10  # 模拟路径数

# 模拟GBM
S = simulate_gbm(S0, mu, sigma, T, dt, N)

# 绘制结果
plt.figure(figsize=(10, 6))
for i in range(N):
    plt.plot(S[i], lw=1)
plt.title('Geometric Brownian Motion Simulation')
plt.xlabel('Time Steps')
plt.ylabel('Stock Price')
plt.show()
```

```python
def gbm_euler(S0, mu, sigma, T, dt, N):
    """
    模拟几何布朗运动
    S0: 初始价格
    mu: 期望收益率
    sigma: 波动率
    T: 总时间
    dt: 时间步长
    N: 模拟路径数
    """
    # 计算步数
    steps = int(T / dt)
    # 初始化价格矩阵
    S = np.zeros((N, steps))
    S[:, 0] = S0  # 设置初始价格

    for i in range(1, steps):
        # 生成标准正态分布随机数
        Z = np.random.normal(0, 1, N)
        # 欧拉方法更新价格
        #S[:, i] = S[:, i - 1] * np.exp((mu - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * Z)
        S[:,i] = S[:,i-1]+mu*S[:,i-1]*dt + sigma * S[:,i-1] * np.sqrt(dt) * Z
    return S
```

```python
# 参数设置
S0 = 100  # 初始价格
mu = 0.1  # 期望收益率
sigma = 0.2  # 波动率
T = 1  # 总时间（年）
dt = 0.01  # 时间步长
N = 10  # 模拟路径数

# 模拟GBM
S = gbm_euler(S0, mu, sigma, T, dt, N)
plt.figure(figsize=(10, 6))
for i in range(N):
    plt.plot(S[i], lw=1)
plt.title('Geometric Brownian Motion Simulation')
plt.xlabel('Time Steps')
plt.ylabel('Stock Price')
plt.show()
```
